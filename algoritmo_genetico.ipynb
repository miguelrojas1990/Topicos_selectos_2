```python
import random
import string

# Definir la cadena objetivo
objetivo = "Grillo malote!"

# Función para generar una cadena aleatoria
def generar_cadena_longitud_aleatoria():
    longitud = len(objetivo)
    return ''.join(random.choice(string.printable) for _ in range(longitud))

# Función de aptitud (fitness) para evaluar cuán cercana está una cadena al objetivo
def aptitud(cadena):
    return sum(1 for a, b in zip(cadena, objetivo) if a == b)

# Función para realizar la selección basada en la aptitud
def seleccion(poblacion):
    return random.choices(poblacion, weights=[aptitud(individuo) for individuo in poblacion], k=len(poblacion))

# Función para realizar la cruza (crossover)
def cruza(padre, madre):
    punto_cruza = random.randint(1, len(padre) - 1)
    hijo = padre[:punto_cruza] + madre[punto_cruza:]
    return hijo

# Función para realizar la mutación
def mutacion(individuo, tasa_mutacion):
    return ''.join(caracter if random.random() > tasa_mutacion else random.choice(string.printable) for caracter in individuo)

```

```python
# Parámetros del algoritmo genético
tamano_poblacion = 120
tasa_mutacion = 0.01
numero_generaciones = 12000

# Inicializar la población aleatoriamente
poblacion = [generar_cadena_longitud_aleatoria() for _ in range(tamano_poblacion)]

# Ciclo principal del algoritmo genético
for generacion in range(numero_generaciones):
    # Realizar selección basada en aptitud
    poblacion = seleccion(poblacion)
    nueva_poblacion = []

    # Realizar cruza y mutación para crear la nueva generación
    while len(nueva_poblacion) < tamano_poblacion:
        padre = random.choice(poblacion)
        madre = random.choice(poblacion)
        hijo = cruza(padre, madre)
        hijo = mutacion(hijo, tasa_mutacion)
        nueva_poblacion.append(hijo)

    # Actualizar la población con la nueva generación
    poblacion = nueva_poblacion

    # Encontrar el mejor individuo en esta generación y mostrar su aptitud
    mejor_individuo = max(poblacion, key=aptitud)
    print(f"Generación {generacion + 1}: {mejor_individuo} (Aptitud: {aptitud(mejor_individuo)})")

# Resultado final
mejor_individuo = max(poblacion, key=aptitud)
print("Resultado Final:", mejor_individuo)
```
